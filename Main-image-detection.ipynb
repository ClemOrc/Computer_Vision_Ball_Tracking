```python
import os
import random
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO("yolov8_volleyball.pt")  

# Define input and output folders
test_image_folder = "./test/images"
output_folder = "./output_images"

# Create output folder if it doesn’t exist
os.makedirs(output_folder, exist_ok=True)

# Ensure test images exist
if os.path.exists(test_image_folder):
    test_image_files = [f for f in os.listdir(test_image_folder) if f.endswith(('.jpg', '.png'))]

    if len(test_image_files) == 0:
        print("No images found in the test folder!")
    else:
        # Select a few random images
        num_samples = min(3, len(test_image_files))
        selected_test_images = random.sample(test_image_files, num_samples)

        # Create a Matplotlib figure
        fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))
        if num_samples == 1:
            axes = [axes]  # Ensure axes is iterable

        for ax, img_file in zip(axes, selected_test_images):
            img_path = os.path.join(test_image_folder, img_file)
            results = model(img_path)  
            img = cv2.imread(img_path)

            if img is None:
                print(f"Error: Unable to load image {img_file}")
                continue

            # Draw bounding boxes
            for result in results:
                for box in result.boxes.xyxy:
                    x1, y1, x2, y2 = map(int, box[:4])
                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 3)  # Green box
                    cv2.putText(img, "Detected", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                                0.6, (0, 255, 0), 2, cv2.LINE_AA)

            # Save the image with bounding boxes
            output_path = os.path.join(output_folder, img_file)
            cv2.imwrite(output_path, img)
            print(f"Saved: {output_path}")

            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            ax.imshow(img)
            ax.axis("off")
            ax.set_title(img_file)

        plt.show()
else:
    print("Error: Test image folder not found!")
```

**Applied tracking on a video**

```python
import cv2
import numpy as np
from collections import deque
from ultralytics import YOLO

model = YOLO("yolov8_volleyball.pt")

video_path = "back_view.mp4"  
output_video_path = video_path + "_output.avi"

cap = cv2.VideoCapture(video_path)

fps = 30  
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_size = (width, height)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, fps, frame_size)

ball_diameter_meters = 0.21  


prev_position = None  
prev_size = None  
speed_buffer = deque(maxlen=fps // 2)
stable_speed = None
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)  

    for result in results:
        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box[:4])
            ball_center = ((x1 + x2) // 2, (y1 + y2) // 2)  
            ball_width_pixels = x2 - x1  

 
            pixels_per_meter = ball_width_pixels / ball_diameter_meters

            speed_mps = 0 

            if prev_position is not None and prev_size is not None:

                pixel_displacement = np.linalg.norm(np.array(ball_center) - np.array(prev_position))
                displacement_meters = pixel_displacement / pixels_per_meter
                speed_mps_motion = displacement_meters * fps  

                size_change_ratio = ball_width_pixels / prev_size  
                if size_change_ratio > 1:
                    depth_speed_mps = (size_change_ratio - 1) * 5  
                else:
                    depth_speed_mps = 0  


                speed_mps = speed_mps_motion + depth_speed_mps
                speed_kph = speed_mps * 3.6  

                speed_buffer.append(speed_mps)

                if len(speed_buffer) == speed_buffer.maxlen:
                    avg_speed_mps = sum(speed_buffer) / len(speed_buffer)
                    avg_speed_kph = avg_speed_mps * 3.6
                    stable_speed = avg_speed_mps, avg_speed_kph

            prev_position = ball_center  
            prev_size = ball_width_pixels 

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 3)  
            cv2.putText(frame, "Volleyball", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                        0.6, (0, 255, 0), 2, cv2.LINE_AA)

    if stable_speed:
        cv2.putText(frame, f"Speed: {stable_speed[0]:.2f} m/s ({stable_speed[1]:.2f} km/h)", 
                    (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("YOLOv8 Volleyball Speed Estimation", frame)
    out.write(frame)

    frame_count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processed video saved as {output_video_path}")
```

**Smoothing volleyball detection and motion filtering**

```python
import cv2
import torch
import csv
import time
import numpy as np
from collections import deque
from ultralytics import YOLO

model = YOLO("yolov8_volleyball.pt")

input_video = "Long volleyball.mp4"
output_video = input_video+"_tracked_output.mp4"

cap = cv2.VideoCapture(input_video)
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_size = (width, height)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video, fourcc, fps, frame_size)

csv_file = "ball_positions.csv"
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["frame_id", "timestamp", "x", "y"])

ball_positions = deque(maxlen=7)
frame_id = 0
start_time = time.time()
last_valid_position = None
max_distance_threshold = 150  # treshold to exclude movement to fast
angle_threshold = 45  # treshold to exclude big changes in angles
missed_frames = 0  # counter not deleting too many frames

def calculate_angle(p1, p2, p3):
    """Calculate the angle between three points (previous, current, next)."""
    if None in (p1, p2, p3):
        return 0

    v1 = np.array(p1) - np.array(p2)
    v2 = np.array(p3) - np.array(p2)
    
    dot_product = np.dot(v1, v2)
    magnitude = np.linalg.norm(v1) * np.linalg.norm(v2)

    if magnitude == 0:
        return 0

    angle_rad = np.arccos(dot_product / magnitude)
    angle_deg = np.degrees(angle_rad)

    return angle_deg

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1
    timestamp = time.time() - start_time 

    results = model(frame)

    detected = False
    new_position = None

    for result in results:
        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box[:4])
            x_center, y_center = (x1 + x2) // 2, (y1 + y2) // 2 
            detected = True

            if last_valid_position is not None:
                dist = np.linalg.norm(
                    np.array([x_center, y_center]) - np.array(last_valid_position)
                )
                
                if dist > max_distance_threshold and missed_frames < 3:
                    print(f"⚠️ Skipping outlier (too far from {last_valid_position})")
                    detected = False
                    break 
            
            missed_frames = 0  
            
            ball_positions.append((x_center, y_center))
            last_valid_position = (x_center, y_center)
            new_position = (x_center, y_center)

            with open(csv_file, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([frame_id, timestamp, x_center, y_center])

    if not detected:
        missed_frames += 1
        if last_valid_position and missed_frames <= 5:
            ball_positions.append(last_valid_position)
        else:
            ball_positions.append(None) 

    if len(ball_positions) >= 3 and missed_frames == 0:
        p1, p2, p3 = ball_positions[-3], ball_positions[-2], ball_positions[-1]
        angle = calculate_angle(p1, p2, p3)
        
        if angle > angle_threshold:
            print(f"🚨 Unrealistic angle detected")
            ball_positions[-1] = ball_positions[-2]

    valid_positions = [pos for pos in ball_positions if pos is not None]
    if valid_positions:
        avg_x = int(np.mean([p[0] for p in valid_positions]))
        avg_y = int(np.mean([p[1] for p in valid_positions]))
    else:
        avg_x, avg_y = None, None

    for i in range(1, len(valid_positions)):
        x1, y1 = valid_positions[i-1]
        x2, y2 = valid_positions[i]
        cv2.line(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

    if avg_x is not None and avg_y is not None:
        cv2.circle(frame, (avg_x, avg_y), 6, (0, 255, 0), -1)

    out.write(frame)
    cv2.imshow("Tracking", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
print(f"Tracking complete. Positions saved to {csv_file}")
```
